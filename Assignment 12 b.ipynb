# Q2. You have to create a function which will accept batsman name as input and outputs a pie chart suggesting percentage of runs scored with scoring shots. Ex- Runs score in 4's-- 32%, runs scored in 6's - 22%, runs scored in 2's -- 10%

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data=pd.read_csv('matches-Copy1.csv')
match=pd.read_csv('deliveries-Copy1.csv')
```

```python
mask1=match.batsman_runs==1 
mask2=match.batsman_runs==2
mask3=match.batsman_runs==3
mask4=match.batsman_runs==4
mask6= match.batsman_runs==6
```

```python
new_dataframe=match[mask1 | mask2 | mask3| mask4| mask6]
```

```python
def stats(batsman):
    bats=batsman
    new_dataframe[new_dataframe['batsman']==batsman].groupby(['batsman_runs'])['batsman_runs'].sum().plot(kind='pie',figsize=(10,8),label=bats,autopct="%0.2f%%")
    #grouped on the basis of the batsman runs (1,2,3,4,6) exculiding zero and a pie chart is created. score sum up to find its percentage on overall scores
    plt.ylabel("Runs")
    plt.title("Batsman And Its Statistics")
    plt.legend()
    return plt.show()
```

```python
stats("V Kohli")
```

```python
stats("CH Gayle")
```
