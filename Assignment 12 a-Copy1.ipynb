Q1. You have to create a function which will take a batsman name as input and plot a 2d plot/line plot depicting his sesaon by season performance(in terms of runs scored per season)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data=pd.read_csv('matches-Copy1.csv')
match=pd.read_csv('deliveries-Copy1.csv')
```

```python
m=data.merge(match,left_on="id",right_on="match_id") # data sheet needed to be merged 
```

```python
def stats(batsman):
    bats=batsman
    m[m['batsman']==batsman].groupby("season")['batsman_runs'].sum().plot(kind='pie',figsize=(10,8),label=bats)
    plt.xlabel("Season")
    plt.ylabel("Runs")
    plt.title("Batsman And Its Statistics")
    plt.legend()
    return plt.show()
```

```python
stats("V Kohli")
```

```python
stats("CH Gayle")
```
